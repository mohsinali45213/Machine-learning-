```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")
df.head()
```

```python
sns.regplot(data = df , x = "total_bill" , y = "tip")
```

```python
# Styling the plot
sns.regplot(data = df , x = "total_bill" , y = "tip",color="red",marker="o"
            ,scatter_kws={"s":60,"color":"green","alpha":0.5}
            ,line_kws={"color":"red","linewidth":2})
```
