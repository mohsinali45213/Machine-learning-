```python
# data (DataFrame or array-like): Data to plot, either a DataFrame or a simple array.
# x (array-like, optional): Data for the x-axis, or column name if data is a DataFrame.
# y (array-like, optional): Data for the y-axis, or column name if data is a DataFrame.
# hue (string, optional): Categorical variable to color the plot.
# multiple ({'layer', 'dodge', 'stack'}, default: 'layer'): How to handle multiple datasets (overlay, side by side, or stack).
# fill (bool, default: True): Whether to fill the area under the KDE curve.
# bw_adjust (float, optional): Adjusts the KDE bandwidth; higher values smooth the curve.
# cut (float, default: 3.0): Extends KDE beyond the data range; higher values for smoother curves.
# thresh (float, optional): Minimum density level to display; values below this are omitted.
# common_norm (bool, default: True): Normalizes all datasets to the same scale.
# ax (matplotlib.axes.Axes, optional): Axes to plot on; if not provided, creates new axes.
# cmap (string or Colormap, optional): Colormap for KDE curves.
# linewidth (float, optional): Width of the KDE curve line.
# legend (bool, default: True): Whether to show a legend for hue variable.
# log_scale (bool or tuple, default: False): Applies a log scale to the axes.
# palette (dict, list, or string, optional): Colors for different levels of hue.
# shade (bool, default: True): Deprecated; use fill instead.
# shade_lowest (bool, default: True): Whether to shade the lowest contour.
# warn_singular (bool, default: True): Warns if the data is singular (e.g., contains duplicates).
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
sns.kdeplot(x = "age",data = df)
```

```python
# kde plot for all numeric variables in Database
sns.kdeplot(data=df)
```

```python
# smoothing using "bw_adjust"
sns.kdeplot(x="age",data=df,bw_adjust=0.5)
```

```python
# Group the KDE on categorical variable
sns.kdeplot(x="age",hue="class",data=df)
```

```python
# Stack KDE on a category using MULTIPLE arg
sns.kdeplot(x="age",hue="class",data=df,multiple="stack")
```

```python
# Use log scalling to map the variable in KDE
sns.kdeplot(x="age",data=df,log_scale=True)
```

```python
# Change style using linewidth,palette,alpha etc
sns.kdeplot(x="age",hue="class",data=df,multiple="stack",linewidth=2,palette="Set2",alpha=0.5)
```

```python
# Create bivariant KDE and group it using "hue" and fill usign "fill"
sns.kdeplot(x="age",y="pclass",hue="sex",data=df,fill=True)
```
