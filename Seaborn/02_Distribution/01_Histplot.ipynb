```python
# data: DataFrame or array-like structure containing the data to be plotted.
# x: Name of the variable on the x-axis (numerical data).
# y: Name of the variable on the y-axis (numerical data) - typically used for plotting a 2D histogram.
# hue: Categorical variable that defines subsets of the data, colored differently.
# bins: Number of bins to use for the histogram or an array defining the bin edges. Default is 10.
# binwidth: Width of each bin. If specified, bins is ignored.
# kde: If True, includes a kernel density estimate (KDE) on top of the histogram. Default is False.
# stat: The statistic to use for the y-axis; options include 'count', 'frequency', 'density', or 'probability'.
# element: Type of plot to use for the histogram bars; options include 'bars', 'step', or 'poly'.
# fill: If True, fills the area under the histogram bars; otherwise, plots outlines.
# palette: Color palette for different levels of the hue variable.
# multipe: Determines how multiple histograms are displayed when using the hue parameter to differentiate between subsets of the data e.g:stack,dodge,fill
# color: Color for all elements when not using the hue variable.
# ax: Matplotlib Axes object to draw the plot onto; otherwise, uses the current Axes.
# orientation: Orientation of the plot; 'vertical' (default) or 'horizontal'.
# log_scale: If True, applies a log scale to the y-axis.
# rug: If True, adds a rug plot along the x-axis showing individual data points.
# line_kws: Additional keyword arguments passed to the underlying matplotlib line function for KDE.
# hist_kws: Additional keyword arguments passed to the underlying matplotlib histogram function.
# kwargs: Additional keyword arguments passed to the underlying matplotlib plotting functions.

# Data Types Used in Histplot
# Numerical Data: Used for the x-axis (and optionally the y-axis) to show the distribution of values.
# Categorical Data: Optional, used with the hue parameter to color-code different subsets.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

df = sns.load_dataset("healthexp")
df.head()
```

```python
plt.figure(figsize=(10,3))
sns.histplot(data =df,x="Spending_USD")
```

```python
# Change binwidth
plt.figure(figsize=(4,3))
sns.histplot(data =df,x="Spending_USD",binwidth=2000)
```

```python
# change number of bin and interval
plt.figure(figsize=(8,3))
plt.xticks(range(0,11859,1000))  
sns.histplot(data =df,x="Spending_USD",binwidth=1000)
```

```python
# Combine with KDE
plt.figure(figsize=(8,3))
plt.xticks(range(0,11859,1000))  
sns.histplot(data =df,x="Spending_USD",binwidth=1000,kde=True)
```

```python
# Use HUE and STACK
plt.figure(figsize=(8,3))
plt.xticks(range(0,11859,1000))  
sns.histplot(data =df,x="Spending_USD",binwidth=1000,hue="Country",multiple="stack")
```

```python
# Make it step/poly plot using ELEMENT arg and change the FILL
plt.figure(figsize=(8,3))
plt.xticks(range(0,11859,1000))  
sns.histplot(data =df,x="Spending_USD",hue="Country",binwidth=1000,element="poly")
```

```python
# Use categorical variable and shrink it
sns.histplot(data=df,x="Country",shrink=0.8,stat="probability")
```

```python
# Create a Bivariate Histogram 
sns.histplot(data=df,x="Year",y="Life_Expectancy")
```
