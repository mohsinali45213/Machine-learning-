```python
# data: The dataset to plot, typically a pandas DataFrame.
# hue: Variable that determines the color of points based on categories.
# hue_order: Specifies the order of levels for the hue variable.
# palette: Defines the color palette for different levels of the hue variable.
# vars: List of variables to include in the plot.
# x_vars: Variables to use on the x-axis of the grid.
# y_vars: Variables to use on the y-axis of the grid.
# kind: Type of plot for the off-diagonal subplots (scatter, kde, reg).
# diag_kind: Type of plot for the diagonal subplots (auto, hist, kde).
# markers: Marker styles for different levels of the hue variable.
# height: Height of each facet in inches.
# aspect: Aspect ratio of each facet to adjust width.
# corner: If True, only plots the lower triangle of the pair grid.
# dropna: If True, removes observations with missing values.
# plot_kws: Additional arguments passed to the plot components.
# diag_kws: Additional arguments for the diagonal plot components.
# grid_kws: Additional arguments passed to the PairGrid class.
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(df)
```

```python
# Change digonal kind to kde,hist or none
sns.pairplot(df,diag_kind="kde")
```

```python
# Change non digonal plot KIND to scatter,kde,hist or reg
sns.pairplot(df,kind="reg")
```

```python
# Add HUE
sns.pairplot(df,hue = "species")
```

```python
# Create pairplot for specific list of variable
sns.pairplot(data=df,x_vars=["sepal_length","sepal_width","petal_length"] , y_vars=["petal_width"])
```

```python
# Showing lower triangle usign CORNER arg
sns.pairplot(data=df,corner=True)
```

```python
# Style both Diagonal  and non diagonal
sns.pairplot(data=df,diag_kws=dict(color="green",kde=True) , plot_kws=dict(color="red",marker="*",s=100))
```

```python
# Create KDE plot on top of the pair plot usign map_lower  and upper
l = sns.pairplot(data=df,plot_kws=dict(color="red"))
l.map_lower(sns.kdeplot)
 
```
