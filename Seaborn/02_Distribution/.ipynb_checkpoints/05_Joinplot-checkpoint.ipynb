```python
# data: DataFrame, array, or list of arrays, optional. Dataset for plotting.
# x: Vector data, optional. Values for the x-axis, which can be a variable name from data or an array-like object.
# y: Vector data, optional. Values for the y-axis, which can be a variable name from data or an array-like object.
# hue: Vector data or key in data, optional. Semantic variable for color mapping.
# kind: String, optional. Type of plot to draw; options are 'scatter' (default), 'kde', 'hist', 'hex', 'reg', or 'resid'.
# stat_func: Callable or None, optional (deprecated). A function to calculate a statistic about the relationship and annotate the plot.
# marginal_kws: Dictionary, optional. Additional keyword arguments for the marginal plots.
# joint_kws: Dictionary, optional. Additional keyword arguments for the joint plot (e.g., markers, sizes).
# hue_order: List of strings, optional. Order for the levels of the hue variable.
# hue_norm: Tuple of floats or Normalize object, optional. Normalization range for scaling the hue variable.
# palette: String, list, dict, or seaborn color palette, optional. Colors to use for mapping the hue variable.
# height: Float, optional. Size of the figure in inches. Default is 6.
# ratio: Float, optional. Ratio of the joint axes height to marginal axes height. Default is 5.
# space: Float, optional. Space between the joint and marginal axes. Default is 0.2.
# dropna: Boolean, optional. If True, drops missing values from the dataset. Default is True.
# xlim: Tuple of floats, optional. Limits for the x-axis.
# ylim: Tuple of floats, optional. Limits for the y-axis.
# marginal_ticks: Boolean, optional. If True, draws ticks on the marginal axes.
# joint_kws: Dictionary, optional. Additional keyword arguments for the plot type used in the joint axes.
# marginal_kws: Dictionary, optional. Additional keyword arguments for the plot type used in the marginal axes.
# corner: Boolean, optional. If True, removes the upper triangle of the plot, including marginal plots.
# fill: Boolean, optional. If True, fills the area under KDE plots.
# legend: Boolean, optional. If True, adds a legend to the plot.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df.head()
```

```python
plt.figure(figsize=(2,2))
sns.jointplot(data=df,x="sepal_length",y="petal_length")
```

```python
# Change its kind options are 'scatter' (default), 'kde', 'hist', 'hex', 'reg', or 'resid'
sns.jointplot(data=df,x="sepal_length",y="petal_length",kind="reg")
```

```python
# Group base on categorical variable
sns.jointplot(data=df,x="sepal_length",y="petal_length",hue="species")
```

```python
# Formating all and Individual plot
sns.jointplot(data=df,x="sepal_length",y="petal_length"
              ,hue="species"
              ,palette="Set2"
              ,joint_kws=dict(marker="+"
                            ,color="red"
                            )
             ,marginal_kws=dict(color="green")
             )
```
