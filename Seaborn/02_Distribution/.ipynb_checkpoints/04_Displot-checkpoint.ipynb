```python
# data: DataFrame, array, or list of arrays, optional. Data to plot.
# x: Vector data, optional. Values for the x-axis. Accepts a variable name from data or an array-like object.
# y: Vector data, optional. Values for the y-axis (mainly used for bivariate plots).
# hue: Vector data or key in data, optional. Semantic variable for color mapping.
# row: Vector data or key in data, optional. Creates subplots across rows.
# col: Vector data or key in data, optional. Creates subplots across columns.
# col_wrap: Integer, optional. Wraps the column facets into multiple rows if the number exceeds this value.
# kind: String, optional. Type of plot to draw; options are 'hist', 'kde', 'ecdf', or 'rug'. Default is 'hist'.
# bins: Integer or sequence, optional. Number of bins for histograms, or bin edges.
# binwidth: Float or array-like, optional. Width of each bin for histograms.
# binrange: Tuple of floats, optional. Lower and upper range of bins.
# discrete: Boolean, optional. If True, bins are treated as discrete values.
# log_scale: Boolean or tuple, optional. If True or (True, False), applies log scaling to the x and/or y axis.
# cumulative: Boolean, optional. If True, plots the cumulative distribution.
# common_bins: Boolean, optional. If True, uses the same bin edges for all facets. Default is True.
# common_norm: Boolean, optional. If True, scales the density axis by the same factor across facets. Default is True.
# multiple: String, optional. Method for multiple distributions on the same plot; options include 'layer', 'dodge', 'stack', or 'fill'.
# element: String, optional. Visual representation of histograms; options are 'bars', 'step', or 'poly'.
# stat: String, optional. Statistic to plot; options are 'count', 'frequency', 'probability', 'percent', or 'density'. Default is 'density'.
# kde: Boolean, optional. If True, adds a KDE plot to the histogram.
# kde_kws: Dictionary, optional. Additional keyword arguments for the KDE plot.
# line_kws: Dictionary, optional. Keyword arguments for line plots when kind is 'kde' or 'ecdf'.
# fill: Boolean, optional. If True, fills the area under the KDE or ECDF curve.
# rug: Boolean, optional. If True, adds a rug plot.
# rug_kws: Dictionary, optional. Additional keyword arguments for the rug plot.
# hue_order: List of strings, optional. Order for the hue levels.
# hue_norm: Tuple of floats or Normalize object, optional. Normalization range for scaling the hue variable.
# legend: Boolean, optional. If True, adds a legend. Default is True.
# palette: String, list, dict, or seaborn color palette, optional. Palette for mapping the hue variable.
# height: Float, optional. Height (in inches) of each facet. Default is 5.
# aspect: Float, optional. Aspect ratio of each facet. Default is 1.
# facet_kws: Dictionary, optional. Additional keyword arguments for FacetGrid.
# ax: matplotlib.axes.Axes, optional. Pre-existing axes for the plot. Otherwise, calls plt.gca() internally.
# kwargs: Additional keyword arguments passed to underlying plotting functions (hist, kde, rug).
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
sns.displot(x="age",data=df,kde=True,rug=True,rug_kws={"height":0.05},kde_kws={"bw_adjust":0.50},hue="sex",multiple="stack",col="alive",row="sex")
```

```python
sns.displot(x="age",data=df,col="alive",row="sex",kind="ecdf")
```
