```python
# data: DataFrame, array, or list of arrays, optional. Data to plot.
# x: Vector data, optional. Values to use for the x-axis. Accepts a variable in data or directly as an array-like object.
# y: Vector data, optional. Values to use for the y-axis. Accepts a variable in data or directly as an array-like object.
# hue: Vector data or key in data, optional. Semantic variable used for mapping color.
# height: Float, optional. Proportion of the axis to use for the rug ticks. Default is 0.025.
# expand_margins: Boolean, optional. If True, expand the axis margins to make room for the ticks. Default is True.
# palette: String, list, dict, or seaborn color palette, optional. Palette to use for mapping the hue variable to colors.
# hue_order: List of strings, optional. Order of levels for the hue variable, affecting the order of colors.
# hue_norm: Tuple of floats or Normalize object, optional. Normalization for scaling the hue variable.
# legend: Boolean, optional. Whether to add a legend for the hue variable. Default is True.
# ax: matplotlib.axes.Axes, optional. Pre-existing axes for the plot. Otherwise, calls plt.gca() internally.
# kwargs: Additional keyword arguments passed to matplotlib.axes.Axes.axvline or axhline, depending on whether x or y is used.

# Purpose: To plot individual data points along an axis as small ticks.
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
# RugPlot for single variable
sns.rugplot(data=df,x="age")
```

```python
# RugPlot for two variable
sns.rugplot(data=df,x="age",y="fare")
```

```python
# Group it by a Categrical variable using hue
sns.rugplot(data=df,x="age",y="fare",hue="sex")
```

```python
# Change the Height 
sns.rugplot(data=df,x="age",y="fare",hue="sex",height=0.1)
```

```python
# Combine RUG with KDE plot
sns.rugplot(data=df,x="age",y="fare",hue="sex",height=0.05)
sns.kdeplot(data=df,x="age",y="fare",hue="sex",fill=True)
```

```python
# Combine RUG with Scatter plot
sns.scatterplot(data=df,x="age",y="fare",hue="sex")
sns.rugplot(data=df,x="age",y="fare",hue="sex",height=0.1)
```

```python
# Show rug outside of axis
sns.rugplot(data=df,x="age",y="fare",hue="sex",height=-0.01,clip_on=False)
```
