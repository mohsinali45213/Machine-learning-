```python
# x: Name of the variable on the x-axis, or a vector of values (categorical data).
# y: Name of the variable on the y-axis, or a vector of values (numerical data).
# hue: Categorical variable that defines subsets of the data, which are colored differently.
# data: DataFrame or array-like structure containing the data for plotting.
# order: Specific order to plot the categorical levels along the x-axis.
# hue_order: Specific order to plot the categorical levels of the hue variable.
# dodge: When True and using a hue variable, separates points along the categorical axis, making subgroups distinct.
# orient: Orientation of the plot; 'v' for vertical (default) or 'h' for horizontal.
# color: Color for all elements or used as a base color when the hue variable is not specified.
# palette: Color palette for different levels of the hue variable.
# size: Size of the markers.
# edgecolor: Color of the marker edges (default is 'gray').
# linewidth: Width of the lines around the markers.
# marker: Type of marker used for the points (default is 'o' for circle).
# alpha: Opacity level of the markers (default is fully opaque).
# ax: Matplotlib Axes object to draw the plot onto, otherwise uses the current Axes.
# warn_thresh: Threshold for number of points above which a warning will be issued when rendering a large number of points.
# kwargs: Additional keyword arguments passed to the underlying matplotlib scatter function.

# Use Case :- same as stripplot
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
df.head()
```

```python
sns.swarmplot(y="age",data=df)
```

```python
# groupby Category
plt.figure(figsize=(10,5))
sns.swarmplot(x="class",y="age",data=df,hue="sex")
```

```python
# Separate hue variable value using "dodge"
plt.figure(figsize=(10,5))
sns.swarmplot(x="class",y="age",data=df,hue="sex",dodge=True)
```

```python
# Change the marker size,color,edge color etc
plt.figure(figsize=(10,5))
sns.swarmplot(x="class",y="age",data=df,hue="sex",marker="*",size=5,palette="Set1",edgecolor="black")
```
