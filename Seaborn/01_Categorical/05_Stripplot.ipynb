```python
# x: Name of the variable on the x-axis, or a vector of values (categorical data).
# y: Name of the variable on the y-axis, or a vector of values (numerical data).
# hue: Categorical variable that defines subsets of the data, colored differently.
# data: DataFrame or array-like, containing the data for plotting.
# order: Specific order to plot the categorical levels along the x-axis.
# hue_order: Specific order to plot the categorical levels of the hue variable.
# jitter: Amount of jitter (spread) to apply along the categorical axis to avoid overlap; can be True, False, or a float specifying the amount.
# dodge: When True and using a hue variable, separates points along the categorical axis.
# orient: Orientation of the plot; 'v' for vertical or 'h' for horizontal.
# color: Color for all elements or a seed for mapping colors with the hue variable.
# palette: Color palette for different levels of the hue variable.
# size: Size of the markers.
# edgecolor: Color of the marker edges.
# linewidth: Width of the lines around the markers.
# marker: Type of marker to use for the points (default is 'o' for circle).
# alpha: Opacity level of the markers (default is fully opaque).
# ax: Matplotlib Axes object to draw the plot onto, otherwise uses the current Axes.
# kwargs: Additional keyword arguments passed to the underlying matplotlib scatter function.

# Categorical Data: Used on the x-axis (or y-axis if oriented horizontally) to define groups.
# Numerical Data: Used on the y-axis (or x-axis if oriented horizontally) to show the distribution of values within each group.
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
df.head()
```

```python
# Basic Stripplot
sns.stripplot(y="age",data =df)
```

```python
# Expand markers using "jitter"
sns.stripplot(y="age",data =df,jitter=0.3)
```

```python
# Draw the line around the point using "linewidth"
sns.stripplot(y="age",data =df,linewidth=1,jitter=0.2)
```

```python
sns.stripplot(x="class",y="age",hue="sex",data =df,linewidth=1,jitter=0.2)
```

```python
# Separate HUE value using "dodge"
sns.stripplot(x="class",y="age",hue="sex",data =df,linewidth=1,jitter=0.2,dodge=True)
```

```python
# Combine Stripplot and Violinplot
sns.stripplot(x="class",y="age",data=df,linewidth=1)
sns.violinplot(x="class",y="age",data=df,linewidth=1)
```

```python
# Combine Stripplot and Boxplot
sns.stripplot(x="class",y="age",data=df,linewidth=1)
sns.boxplot(x="class",y="age",data=df,linewidth=1)
```
