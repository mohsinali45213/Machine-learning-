```python
# x: Name of the variable on the x-axis, or a vector of values (categorical data).
# y: Name of the variable on the y-axis, or a vector of values (numerical data).
# hue: Categorical variable that will determine the color of the violins.
# data: DataFrame or array-like, containing the data for plotting.
# order: Specific order to plot the categorical levels along the x-axis.
# hue_order: Specific order to plot the categorical levels of the hue variable.
# bw_method: Bandwidth of the kernel density estimation, affecting the smoothness of the density curve (default is 0.2).
# cut: Extends the density past the extreme datapoints by this number of standard deviations (default is 2).
# scale: Method to scale the width of each violin. Options are 'area' (default), 'count', or 'width'.
# scale_hue: If True, the width of the violins will be scaled by the hue variable.
# gridsize: Number of points to evaluate the density at, affects smoothness (default is 100).
# width: Width of each violin (default is 0.8).
# inner: Representation of the data points in the violin: 'box', 'quartile', 'point', 'stick', or None.
# split: When using a hue variable, splits the violins in half to compare the distribution across hue levels.
# dodge: When using a hue variable, separates the violins along the categorical axis.
# linewidth: Width of the lines used for the violin edges.
# color: Color for all elements, or seed for mapping colors with the hue variable.
# palette: Color palette for different levels of the hue variable.
# saturation: Proportion of the original saturation to be applied to the colors (default is 0.75).
# ax: Matplotlib Axes object to draw the plot onto, otherwise uses the current Axes.
# kwargs: Additional keyword arguments passed to the underlying matplotlib function

# Use Case :-
# Numerical Data::- This provide summary of supplied data like :-Medium,Minimum,Maximum,1st Quartile,3rd Quartile,KDE(Kernel Density Estimation),
# Categorical Data: Used on the x-axis to split the numerical data into different groups or categories.
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# Violineplot use on one numerical variable
sns.violinplot(y="total_bill" ,data=df)
```

```python
# Violineplot on categorical and numerical variables and use Split
sns.violinplot(x="time",y="total_bill",hue="sex",data=df,split=True)
```

```python
# Change box direction in the Violinplot
sns.violinplot(x="time",y="total_bill",hue="sex",data=df,split=True,inner="quartile")
```

```python
# Drow line for each observations
sns.violinplot(x="time",y="total_bill",hue="sex",data=df,split=True,inner="stick")
```

```python
# Change the amount of smooth using "bw_method"
sns.violinplot(x="time",y="total_bill",hue="sex",data=df,split=True,inner="quartile",bw_method=1)
```
