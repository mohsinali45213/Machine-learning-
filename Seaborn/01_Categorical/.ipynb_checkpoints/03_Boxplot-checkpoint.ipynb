```python
# x: Name of the variable on the x-axis, or a vector of values.
# y: Name of the variable on the y-axis, or a vector of values.
# hue: Variable that defines subsets of the data, which will be colored differently.
# data: DataFrame or array-like structure that holds the data.
# order: Specific order to plot the categorical levels along the x-axis.
# hue_order: Specific order to plot the categorical levels of the hue variable.
# orient: 'v' for vertical, 'h' for horizontal orientation.
# color: Color for all elements or to be used when the hue variable is not specified.
# palette: Color palette for different levels of the hue variable.
# saturation: Proportion of the original saturation to be applied to the colors.
# width: Width of the boxes in the plot (default is 0.8).
# dodge: When True, separates boxes by hue along the categorical axis.
# fliersize: Size of the markers used for outliers.
# linewidth: Width of the lines used for boxes, whiskers, caps, and medians.
# whis: Proportion of the IQR past the first and third quartiles to extend the whiskers. It can also be set as a float or a tuple.
# notch: If True, produces a notched box plot to indicate confidence intervals around the median.
# ax: Matplotlib Axes object to draw the plot onto, otherwise uses the current Axes.
# kwargs: Additional keyword arguments are passed directly to the underlying matplotlib boxplot function.
# showmeans: If True, displays the mean of the data as a point.
# meanline: If True, will show the mean as a line instead of a point.
# showbox: If False, removes the boxes but retains the whiskers.
# showfliers: If False, suppresses the display of outliers.
# showcaps: If False, suppresses the display of the caps at the end of whiskers.
# boxprops: Dictionary of properties for the boxes.
# whiskerprops: Dictionary of properties for the whiskers.
# capprops: Dictionary of properties for the caps.
# flierprops: Dictionary of properties for the outliers (fliers).
# medianprops: Dictionary of properties for the median line.
# meanprops: Dictionary of properties for the mean line or point.
# ci: Not applicable in Seaborn boxplot; sometimes seen as a general parameter in Seaborn functions but not directly used here.

# Use Case :-
# Numerical Data::- This provide summary of supplied data like :- Mean,Medium,Minimum,Maximum,1st Quartile,3rd Quartile,Outliers
# Categorical Data: Used on the x-axis to split the numerical data into different groups or categories.
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# Basic BOX plot on one Numrical variable 
sns.boxplot(y="total_bill" , data=df ,width=0.2)
```

```python
# Boxplot using numrical and categorical value with HUE
sns.boxplot(x ="time",hue = "sex",y="total_bill",data=df )
```

```python
# showmeans=True is use to mark mean value in boxplot and meanprops use to change marker style
sns.boxplot(x ="time",hue = "sex",y="total_bill",data=df,showmeans=True,meanprops ={"marker":"o",
                                                                                    "markerfacecolor":"white",
                                                                                    "markeredgecolor":"black"
                                                                                   })
```

```python
# change palette and line width
sns.boxplot(x ="time",hue = "sex",y="total_bill",data=df,palette="Set2",linewidth=4)
```

```python
# Box plot for each of the numerical variable in the dataframe
sns.boxplot(data=df)
```
