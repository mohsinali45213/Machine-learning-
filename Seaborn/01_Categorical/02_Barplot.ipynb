```python
# x: The name of the variable in data for the x-axis (categorical).
# y: The name of the variable in data for the y-axis (numerical).
# hue: The name of the variable in data to color the bars (optional).
# data: The dataset (usually a DataFrame) containing the variables.
# palette: Color palette for the bars (e.g., 'viridis', 'Blues', 'Set1').
# order: Order of categories on the x-axis (list of category names).
# hue_order: Order of the hue categories (list of hue names).
# ci,errorbar: Confidence interval for error bars (default is 95; None to disable).
# estimator: Function to aggregate data (e.g., mean, sum, or a custom function).
# dodge: Whether to dodge bars for hue levels (default is True).
# orient: Orientation of the plot ('v' for vertical, 'h' for horizontal).
# errcolor: Color of the error bars (if ci is not None).
# errwidth: Width of the error bars (if ci is not None).
# capsize: Size of the caps on the error bars (if ci is not None).
# linewidth: Width of the line for the bars.
# saturation: Proportion of the saturation for the colors in the palette (default is 0.75).

# Use Case: To compare quantities of different categories or groups.
# Data Type: It typically uses categorical data on the x-axis and numerical data on the y-axis.
```

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# sns.get_dataset_names()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x = "day",y = "total_bill", hue="sex" , data = df)
```

```python
# Add cap on error bar
plt.figure(figsize=(10,4))
sns.barplot(x = "day",y = "total_bill", hue="sex" , capsize=0.10, data = df)
```

```python
# Remove error bar using "ci"
plt.figure(figsize=(10,4))
sns.barplot(x = "day",y = "total_bill", hue="sex" ,errorbar=None, data = df)
```

```python
# Default estimator is mean
plt.figure(figsize=(10,4))
sns.barplot(x = "day",y = "total_bill", hue="sex" ,errorbar=None, data = df,estimator="sum")
```
