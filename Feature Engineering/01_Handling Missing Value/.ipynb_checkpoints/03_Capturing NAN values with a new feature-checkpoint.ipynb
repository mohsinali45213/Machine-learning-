#### <b><code>3. Capturing NAN values with a new feature</code></b>
<pre>It works well if the data are not missing completely at random</pre>

```python
import pandas as pd 
import numpy as np

df=pd.read_csv('../titanic.csv', usecols=['Age','Fare','Survived'])
df.head()
```

```python
df["Age_NUN"] = np.where(df["Age"].isnull(),1,0)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df.head(20)
```

#### <b><code>Advantages</code></b>
<pre>1. Easy to implement
2. Captures the importance of missing values</pre>
#### <b><code>Disadvantages</code></b>
<pre>1. Creating Additional Features(Curse of Dimensionality)</pre>
