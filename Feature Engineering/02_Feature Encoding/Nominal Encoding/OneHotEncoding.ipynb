```python
df = pd.read_csv("../titanic.csv",usecols=["Cabin"])
df.head()
```

```python
df.dropna(inplace=True)
df.Cabin.unique()
```

```python
df.Cabin.value_counts().sort_values(ascending=False).head(10)
```

```python
last_10 = df.Cabin.value_counts().sort_values(ascending=False).head(10).index
last_10
```

```python
for cat in last_10:
    df[cat] = np.where(df['Cabin']==cat,1,0)
```
