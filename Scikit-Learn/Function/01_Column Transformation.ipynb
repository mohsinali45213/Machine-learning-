```python

import pandas as pd 
import seaborn as sns
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
new_df = df.iloc[:,[3,2,8,9,14,13]]
new_df
```

```python
new_df.isnull().sum()
```

```python
transformation = ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),["age"]),
    ("tnf2",OrdinalEncoder(categories=[["man","woman","child"],["First","Second","Third"]]),["who","class"]),
    ("tnf3",OneHotEncoder(drop="first",sparse_output=False),["sex","alone","alive"])
])
```

```python
new_df = pd.DataFrame(transformation.fit_transform(new_df),columns=["age","who","class","sex","alone","alive"]).head(50)
```
