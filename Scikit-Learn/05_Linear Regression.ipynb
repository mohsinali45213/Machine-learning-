```python
# Linear Regression
# Linear Regression deals with continues value ,A continuous value is a number that can take any value within a range
# steps:-
# 1 We ploat our dependent variable (Y-axis) and independent variable (X-axis)
# 2 We try to plot straight line and measure correlation
# 3 We keep changing the direction of our straight line until we get the best correlation 
# 4 We extrapolate from this line to find new value of Y-axis
```

```python
from sklearn import datasets, linear_model,model_selection
import matplotlib.pyplot as plt
```

```python
x,y = datasets.load_iris(return_X_y = True)
x.shape,y.shape
```

```python
plt.scatter(x[:,0],y)
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size = 0.2)
```

```python
linear = linear_model.LinearRegression()
linear.fit(x_train,y_train)

pred = linear.predict(x_test)
```

```python
plt.scatter(pred,y_test)
```

```python
linear.score(x_test,y_test)
```
