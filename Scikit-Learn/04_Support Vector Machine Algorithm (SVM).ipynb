```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x =  iris.data
y = iris.target
names = iris.target_names
```

```python
df = pd.DataFrame(x,columns=iris.feature_names)
df["species"] = iris.target
df["species"] = df["species"].replace(to_replace=[0,1,2], value=["setpsa","versicolor","virginica"])
df.head()
```

```python
sns.pairplot(data = df , hue = "species" , palette = "Set2")
```

```python
# Segregate our data for training and testing 
#test_size = 0.20 represent 20% of data split into train 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Apply KNN algorithm and predict x_test data
from sklearn.svm import SVC
svm = SVC(kernel="linear",random_state=0)
svm.fit(x_train,y_train)
```

```python
pred = svm.predict(x_test)
pred
```

```python
# find the accuracy of y_test base on x_test prediction 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
accuracy_score(y_test,pred)
```
