```python
# Data Cleanign and Preprocessing
import pandas as pd
from sklearn.datasets import fetch_openml
```

```python
#data importing from sklearn
df = fetch_openml("titanic",version=1,as_frame=True)["data"]
```

```python
# finding null value from data base 
df.isnull().sum()
```

```python
# graphical representation 
import seaborn as sns
```

```python
sns.set()
```

```python
# finding null values in percentage(%) of each row
miss_val_per = pd.DataFrame((df.isnull().sum()/len(df))*100)
miss_val_per.plot(kind="bar" , title="Missing Value" , ylabel="Percentage")
```

```python
# size of the dataset
# (row,column)
df.shape
```

```python
# drop column from the database
# axis=1 represented column and axis=0 represented row
# inplace=True use to do permanent changing in orignal database
df.drop(['body'],axis=1,inplace=True)
```

```python
# size of the database after drop the column
df.shape
```

```python
# Value Imputation
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
# total number of null value in age column befor imputing
int(df["age"].isnull().sum())
```

```python
# total number of null value in age column after imputing
imp = SimpleImputer(strategy="mean")
df["age"] = imp.fit_transform(df[["age"]])
int(df["age"].isnull().sum())
```

```python
#fill null value in holl database
# this function use to get all column which contain null balue and staturgy base on datatype
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
       if df[col].dtype in ["float64", "int64", "int32"]:
           strategy = "mean"
       else:
           strategy = "most_frequent"
       missing_values = df[col][df[col].isnull()].values[0]
       parameters[col] = {"missing_values":missing_values,"strategy":strategy}
    return parameters
get_parameters(df)
```

```python
parameter = get_parameters(df)
```

```python
#imputing all null value in database
for col,param in parameter.items():
    missing_values = param["missing_values"]
    strategy = param["strategy"]
    imp = SimpleImputer(missing_values=missing_values,strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```
